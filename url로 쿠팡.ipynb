```python
html = bs(response.text, "lxml")
```

```python

review = html.find_all("div", {"class": "sdp-review__article__list__review__content js_reviewArticleContent"})
answer = html.find_all("div", {"class": "sdp-review__article__list__survey"})
date=html.find_all("div", {"class": "sdp-review__article__list__info__product-info__reg-date"})
score=html.find_all("div", {"class": "sdp-review__article__list__info__product-info__star-orange js_reviewArticleRatingValue"})
```

```python
a=date[0].get_text().replace('\n','').strip()
b=review[0].get_text().replace('\n','').strip()
ans_sound=answer[0].get_text().replace('\n','').strip().replace('디자인','').replace('착용감','-').replace('음질','-').split('-')[0]
ans_fit=answer[0].get_text().replace('\n','').strip().replace('디자인','').replace('착용감','-').replace('음질','-').split('-')[1]
ans_design=answer[0].get_text().replace('\n','').strip().replace('디자인','').replace('착용감','-').replace('음질','-').split('-')[2]
star=score[0].get("data-rating")
```

```python
data={'date':[a],'review':[b],'sound':[ans_sound],'fit':[ans_fit],'design':[ans_design],'score':[star]}
```
